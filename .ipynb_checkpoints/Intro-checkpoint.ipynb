**Data Description:** This is a real dataset for a UK company selling unusual gifts. It contains all transactions that occurred between December 1, 2009, and December 9, 2011.

**Objective:** To understand store performance, customer behavior, and improve sales.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Reading data from a CSV file

```python
df = pd.read_csv('Data/data.csv', encoding='unicode_escape')
```

Displaying basic information about the data

```python
print(df.info())
print(df.head())
```

**Data cleaning**

```python
# 1. Remove rows containing an empty CustomerID (for customer analysis)
df_clean = df.dropna(subset=['CustomerID'])

# 2. Check for any negative values â€‹â€‹in quantity or price (these may be returns or errors)
print(df_clean[(df_clean['Quantity'] <= 0) | (df_clean['UnitPrice'] <= 0)])

# Create a clear copy of the DataFrame
df_clean = df.copy()

# 3. Create a new column for revenue (Quantity * UnitPrice)
df_clean['Revenue'] = df_clean['Quantity'] * df_clean['UnitPrice']

# 4. Convert InvoiceDate to Date and Time type
df_clean['InvoiceDate'] = pd.to_datetime(df_clean['InvoiceDate'])

# 5. Extracting the month and year for time analysis
df_clean['YearMonth'] = df_clean['InvoiceDate'].dt.to_period('M')

print("Data after cleaning:")
print(df_clean.info())
```

**Exploratory data analysis**

**1.** Total monthly revenue (time series analysis)

```python
monthly_revenue = df_clean.groupby('YearMonth')['Revenue'].sum().reset_index()
plt.figure(figsize=(14,6))
plt.plot(monthly_revenue['YearMonth'].astype(str), monthly_revenue['Revenue'])
plt.title('Monthly revenue of the store')
plt.xlabel('month')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

**2.** Top 10 customers in terms of spending (customer value analysis)

```python
top_customers = df_clean.groupby('CustomerID')['Revenue'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_customers.values, y=top_customers.index.astype(str))
plt.title('Top 10 customers in terms of total spending')
plt.xlabel('Total Spending')
plt.show()
```

**3.** Sales distribution by country (pie chart)

```python
country_revenue = df_clean.groupby('Country')['Revenue'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,12))
plt.pie(country_revenue.values, labels=country_revenue.index, autopct='%1.1f%%')
plt.title('Revenue breakdown by country (Top 10)')
plt.show()
```

**4.** Is there a relationship between product price and sales volume? (Scattered diagram)

```python
plt.figure(figsize=(10,6))
plt.scatter(df_clean['UnitPrice'], df_clean['Quantity'], alpha=0.5)
plt.title('The relationship between unit price and quantity sold')
plt.xlabel('unit price')
plt.ylabel('Quantity')
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.show()
```

**Export clean data from Python to Excel**

```python
# Export clean DataFrame to Excel file
df_clean.to_excel('Data/cleaned_ecommerce_data.xlsx', index=False)
```
